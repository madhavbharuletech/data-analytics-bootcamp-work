```python
"""
PYTHON LEARNING MASTER FILE
Author: Madhav S. Bharule

Description:
This file combines all my Python learning exercises,
classroom tasks, and practice problems.

It shows my progression from fundamentals to
data handling and problem solving.
"""

# =========================================================
# SECTION 1 — VARIABLES & TYPES
# =========================================================
print("\n===== VARIABLES & TYPES =====")

myName = "Madhav"
print("Name:", myName)
print("Type:", type(myName))
print("Length:", len(myName))


# =========================================================
# SECTION 2 — AGE CALCULATION
# =========================================================
print("\n===== AGE CALCULATION =====")

from datetime import date

birthdate = date(1990, 5, 25)
today = date.today()

age = today.year - birthdate.year - ((today.month, today.day) < (birthdate.month, birthdate.day))
print("Age:", age)


# =========================================================
# SECTION 3 — CONDITIONAL LOGIC
# =========================================================
print("\n===== CONDITIONAL LOGIC =====")

number = -5

if number > 0:
    print("Positive")
elif number < 0:
    print("Negative")
else:
    print("Zero")


# =========================================================
# SECTION 4 — FIZZBUZZ
# =========================================================
print("\n===== FIZZBUZZ (1–20 SAMPLE) =====")

for i in range(1, 21):
    if i % 15 == 0:
        print("fizzbuzz")
    elif i % 3 == 0:
        print("fizz")
    elif i % 5 == 0:
        print("buzz")
    else:
        print(i)


# =========================================================
# SECTION 5 — LIST OPERATIONS
# =========================================================
print("\n===== LIST OPERATIONS =====")

list1 = ["USA", "China", "India"]
list2 = ["Germany", "Japan", "France"]

list1.extend(list2)
print("Combined:", list1)

list1.pop(2)
print("Removed third element:", list1)

list1.remove("Germany")
print("Removed Germany:", list1)

list1.clear()
print("Cleared list:", list1)


# =========================================================
# SECTION 6 — SETS PRACTICE
# =========================================================
print("\n===== SETS =====")

learner_skills = {"sql", "excel", "python"}

if {"sql", "excel", "python"}.issubset(learner_skills):
    print("Expert Data Analyst")
elif {"sql", "excel"}.issubset(learner_skills):
    print("Good Data Analyst")
else:
    print("Still learning")

europe = {"UK", "Norway", "Switzerland", "France"}
eu = {"France"}
print("Non EU countries:", europe - eu)


# =========================================================
# SECTION 7 — DICTIONARIES
# =========================================================
print("\n===== DICTIONARY BASICS =====")

student = {
    "name": "Madhav",
    "age": 36,
    "city": "Newcastle",
    "course": "Data Analytics"
}

print("Name:", student["name"])

student["age"] = 37
print("Updated:", student)

new_dict = {}
new_dict["language"] = "Python"
print("New dictionary:", new_dict)


print("\n===== CAR DICTIONARY =====")

car = {"brand": "Toyota", "model": "Corolla", "year": 2024}
car["color"] = "White"

for key, value in car.items():
    print(f"{key} : {value}")


print("\n===== DICTIONARY WITH LIST =====")

items = ["Laptop", "Mouse", "Keyboard", "Monitor"]

student2 = {"name": "Madhav", "course": "Python"}
student2["items"] = items

print("Second item:", student2["items"][1])
print("Keys:", student2.keys())
print("Values:", student2.values())

for key, value in student2.items():
    print(f"{key} : {value}")


# =========================================================
# SECTION 8 — WHILE LOOP
# =========================================================
print("\n===== WHILE LOOP =====")

num = 1
while num <= 20:
    print(num, end=" ")
    num += 1
print()


# =========================================================
# SECTION 9 — EVEN NUMBERS
# =========================================================
print("\n===== EVEN NUMBERS =====")

for num in range(0, 101, 2):
    print(num, end=" ")
print()


# =========================================================
# SECTION 10 — PANDAS INTRODUCTION
# =========================================================
print("\n===== PANDAS SECTION =====")

import pandas as pd

df = pd.read_csv("student.csv")

print("\n--- FIRST ROWS ---")
print(df.head())

print("\n--- INFO ---")
print(df.info())

print("\n--- DESCRIBE ---")
print(df.describe())


# =========================================================
# SECTION 11 — FILTERING
# =========================================================
print("\n===== FILTERING =====")

print(df[df['class'] == "Four"])
print(df[(df['class'] == "Four") & (df['gender'] == "Male")])
print(df.query("mark > 75"))


# =========================================================
# SECTION 12 — DATA MANIPULATION
# =========================================================
print("\n===== DATA MANIPULATION =====")

df['passed'] = df['mark'] >= 60
print(df.head())


# =========================================================
# SECTION 13 — GROUPING
# =========================================================
print("\n===== GROUPING =====")

print(df.groupby('class')['mark'].mean())
print(df.groupby('gender')['mark'].mean())


# =========================================================
# SECTION 14 — PIVOT TABLE
# =========================================================
print("\n===== PIVOT TABLE =====")

pivot = df.pivot_table(values='mark', index='class', columns='gender')
print(pivot)


# =========================================================
# SECTION 15 — NULL VALUES
# =========================================================
print("\n===== NULL VALUES =====")

print(df.isnull().sum())
print("Total nulls:", df.isnull().sum().sum())


# =========================================================
# SECTION 16 — FEATURE ENGINEERING
# =========================================================
print("\n===== CREATE GRADES =====")

df['grade'] = df['mark'].apply(
    lambda x: 'A' if x >= 85 else
              'B' if x >= 70 else
              'C' if x >= 60 else
              'D'
)

print(df[['name', 'mark', 'grade']])


# =========================================================
# SECTION 17 — SORTING
# =========================================================
print("\n===== SORTING =====")

sorted_df = df.sort_values(by='mark', ascending=False)
print(sorted_df[['name', 'mark']])


# =========================================================
# SECTION 18 — EXPORT DATA
# =========================================================
print("\n===== SAVE FILE =====")

df.to_csv("students_with_grades.csv", index=False)
print("File saved successfully")


# =========================================================
# FINAL REFLECTION
# =========================================================
print("\n===== LEARNING REFLECTION =====")

print("""
This master file combines all my learning exercises.

It demonstrates:
✔ Python basics
✔ Loops and logic
✔ Data structures
✔ Problem solving
✔ Pandas data analysis

This file represents my learning journey as I build
skills in Python and data analytics.
""")
```
